```python
import pandas as pd
```

```python
#Importing csv as dataframe
df= pd.read_csv("wine_filtered.csv")
df
```

```python
#Number of wines per country
df.pivot_table(index=['country'], aggfunc='size')
```

```python
#Number of wines availed by country
import matplotlib.pyplot as plt

import seaborn as sns

plt.figure(figsize=(14,6))

country = df.country.value_counts()[:20]

g = sns.countplot(x='country', data=df[df.country.isin(country.index.values)])
g.set_title("Country Of Wine Origin Count", fontsize=20)
g.set_xlabel("Country's ", fontsize=15)
g.set_ylabel("Count", fontsize=15)
g.set_xticklabels(g.get_xticklabels(),rotation=45)
```

```python
#Filtering countries with less than 2 wines
dfindex=df.set_index('country')
dfindex
dff=dfindex.drop(['China', 'Egypt', 'Slovakia'] )
dff.pivot_table(index=['country'], aggfunc='size')
dfg=dff.reset_index()
dfg

```

```python
#Ploting the relation points by Country
import matplotlib.pyplot as plt

import seaborn as sns

plt.figure(figsize=(50,20))

plt.subplot(2,1,2)
g1 = sns.boxplot(x='country', y='points',data=dfg)
g1.set_title("Points by Country Of Wine Origin", fontsize=20)
g1.set_xlabel("Country's ", fontsize=15)
g1.set_ylabel("Points", fontsize=15)
g1.set_xticklabels(g1.get_xticklabels(),rotation=70)

plt.subplots_adjust(hspace = 0.6,top = 0.9)

plt.show()
```

```python
#Price by country
import matplotlib.pyplot as plt

import seaborn as sns
plt.figure(figsize=(25,50))

plt.subplot(5,2,1)
g = sns.barplot(x='price', y='country',
                  data=df)
g.set_title("Price by Country Of Wine Origin", fontsize=20)
g.set_xlabel("Country's ", fontsize=15)
g.set_ylabel("Price Dist(US)", fontsize=15)
g.set_xticklabels(g.get_xticklabels(),rotation=70)
```
